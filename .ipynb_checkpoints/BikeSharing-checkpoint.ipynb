Data Analysis

```python
# Trend of Sales by Year and Weekday (2019 has seen more sales compared to 2018)
tempdf = pd.pivot_table(data=df,index=['weekday'],columns='yr',values='cnt',aggfunc='sum')
tempdataframe = pd.DataFrame({'Weekday': tempdf.index,'2018':tempdf[0],'2019':tempdf[1]})
tempdataframe.plot(x="Weekday", y=["2018", "2019"], kind="bar", rot=0)
plt.show()
```
